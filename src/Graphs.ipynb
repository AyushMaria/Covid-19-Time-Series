```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
## Loading the results for the first query.

query_1 = pd.read_csv("query1_monthly_mean_cases.csv")
query_1.head()
```

```python
## Dropping the redundant index column.
query_1 = query_1.drop(["Unnamed: 0"],axis=1)
```

```python
#Plotting the bar plot for the top 10 countries for every month in the dataset.
for col in query_1.columns[1:]:
    sort_df = query_1.sort_values(by=[col],ascending=False)[:10]
    plt.figure(figsize=[16,10])
    sns.barplot(x="Country", y= col,data=sort_df)
    
    
    
    
    
```

## Graph for 2nd query

```python
# Loading the dataset for the second query.
mean_df = pd.read_csv("mean_df.csv")
mean_df.head()
```

```python
#Plotting the line plot for each continent in the dataset, we have taken weeks on the x-axis on a logarithmic scale.
for col in mean_df.columns[1:]:
    plt.subplots(1,1)
    splot = sns.lineplot(data = mean_df,color='green', x= "Weeks", y = col)
    splot.set(xscale="log")
```

## 3rd query

```python
query_3 = pd.read_csv("Query_3_monthly_clusters.csv")
query_3.head()
```

```python
query_3.info()
```

```python
# We plotted the bar plot for top 10 countries to determine the highest trendline coefficient monthly
sort_df= query_3.sort_values(by=["Coefficient"],ascending=False)[:10]
plt.figure(figsize=[16,10])
sns.barplot(x="States_New", y= "Coefficient",data=sort_df)
```

```python
# We plot the category plot to identify clusters for each month in the dataset for the top 10 countries in the dataset.
for col in sort_df.columns[2:]:
    plt.figure(figsize=[16,10])
    g = sns.catplot(x=col, y="States_New", data=sort_df)
    g.set(xticks=[0,1,2,3])

```
